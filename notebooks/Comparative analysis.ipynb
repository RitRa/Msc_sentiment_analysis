```python
df["Category"] = df["Category"].astype("category")
df["Metric"] = df["Metric"].astype("category")
df["Database"] = df["Database"].astype("category")
#df["operations"] = df["operations"].astype("category")
```

# Visualisation of results

```python
def styles():
    sns.set();
    sns.set(font_scale=1.5);
    #plot size
    plt.figure(figsize=(16, 5)); 
```

```python
#call styles for plots
styles();

#select metrics runtime
df_runtime = df.query("Metric == 'RunTime(ms)'")
#chart
ax = sns.pointplot(data=df_runtime, x="Operations", y="Results", hue="Database", palette="viridis");

ax.set_ylabel("RunTime(ms)");
#title
plt.title("RunTime(ms) comparison");
```

```python
#call styles for plots
styles();

#select metrics runtime
df_throughput = df.query("Metric == 'Throughput(ops/sec)'")
#chart
ax = sns.pointplot(data=df_throughput, x="Operations", y="Results", hue="Database", palette="viridis");

ax.set_ylabel("Throughput(ops/sec)");
#title
plt.title("Throughput(ops/sec) comparison");
```

"Throughput is a measure of how many units of information a system can process in a given amount of time." MongoDB can process more units than MySQL

```python
#call styles for plots
styles();

#select metrics runtime
df_95latency = df.query("Metric == '95thPercentileLatency(us)' and Category == 'INSERT'")
#chart
ax = sns.pointplot(data=df_95latency, x="Operations", y="Results", hue="Database" , palette="viridis");

ax.set_ylabel("Latency(us)");
#title
plt.title("Insert: 95thPercentile Latency(us) comparison");
```

"Latency is the total amount of time that it will take for the database to receive a request, process the underlying transaction, and return the correct response." Mongodb wins

```python
#call styles for plots
styles();

#select metrics runtime
df_latency = df.query("Metric == ['95thPercentileLatency(us)', '99thPercentileLatency(us)']")
df_latency= df_latency.query("Operations ==10000 and Category == 'INSERT'")

sns.barplot(data= df_latency, x='Metric', y ='Results' , hue="Database", 
            palette="viridis", order=['95thPercentileLatency(us)', '99thPercentileLatency(us)']);

# y label
ax.set_ylabel("Latency(us)");

# title
plt.title("Insert: 95thPercentile Latency(us) comparison for 10,000 operations", fontsize=14);
```

MongoDB wins
