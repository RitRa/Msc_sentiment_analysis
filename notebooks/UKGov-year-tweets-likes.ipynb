```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from time import sleep
from datetime import datetime
import os

#forma
tweets_list1 = []
e = 0
max_result = 10

#lista = ["2021-05-01", "2021-05-02","2021-05-03", "2021-05-04" ]

from datetime import date, timedelta

# create a list
lista = []

# start date 
start_date = date(2021, 10, 23)
#end date
end_date = date(2022, 10, 23)
delta = timedelta(days=1)
while start_date <= end_date:
    lista.append(start_date.strftime("%Y-%m-%d"))
    start_date += delta
#print(lista)
       
# get tweets 
for y in lista:
    naming = 'ukgov_23'
    term = '(ukgov) until:' +y+ ' lang:en'
    print(naming + " in progress!")
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{term}').get_items()):  # declare a username
        e += 1
        print("Current scrape index: {0} [%d%%]\r".format(e), end="")
        if i>50:
            break
            # print(movie_name, i, tweet)
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.retweetCount, tweet.likeCount, tweet.hashtags ])  # declare the attributes to be returned

    tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username','retweetCount','likeCount','hashtags'])
    tweets_df1.to_csv("{}.csv".format(naming))
```
